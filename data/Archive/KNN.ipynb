```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('../data/fin_final_data.csv')
df.sort_values(by=['date', 'permno'], inplace=True)
df = df.drop(columns=['permno', 'ticker', 'sic2'])
# df = df.drop(columns=['permno', 'Unnamed: 0'])
df
```

```python
# Ensure the 'date' column is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Extract the year as a string
df['year'] = df['date'].dt.strftime('%Y')  # Keeps year as a string

df = df[df['date'] >= '1987-01-01']
df
```

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import ParameterSampler

# Define the training and testing periods
testing_start_date = '2012-01-01'
testing_end_date = '2022-01-01'  # non-inclusive

# Store results
results = []

# Define simplified parameter grid
param_grid = {
    'n_neighbors': [5, 10, 20, 30],  # Number of neighbors
    'weights': ['uniform', 'distance'],  # Weighting strategy
    'p': [1, 2]  # 1 for Manhattan, 2 for Euclidean
}

# Get the testing data
test_data = df[(df['date'] >= testing_start_date) & (df['date'] < testing_end_date)]

# Create a date range for testing with year-end frequency
year_range = pd.date_range(start=testing_start_date, end=testing_end_date, freq='YE')
year_dates = year_range.strftime('%Y-01-01').tolist()

# Select columns excluding target and date columns
selected_columns = [col for col in df.columns if col not in ['date', 'year', 'Risk_Premium']]

# Initialize best parameters with defaults
best_params = {
    'n_neighbors': 5,
    'weights': 'uniform',
    'p': 1
}

# Process each year in the testing period
for year in year_dates:
    print(f"Performing hyperparameter tuning for {year}")
    
    best_score = float('inf')
    
    # Sample a subset of parameter combinations for efficiency
    param_list = list(ParameterSampler(param_grid, n_iter=16, random_state=42))
    
    for params in param_list:
        # Calculate CV score across multiple years
        cv_scores = []
        
        cvyear_range = pd.date_range(start=pd.to_datetime(year) - pd.DateOffset(years=5), 
                                     end=pd.to_datetime(year) - pd.Timedelta(days=1), 
                                     freq='YE')
        cvyear_dates = cvyear_range.strftime('%Y-01-01').tolist()
        
        for cv_year in cvyear_dates:  # Use the last 5 years for CV
            cv_year_dt = pd.to_datetime(cv_year)
            new_start_date = cv_year_dt - pd.DateOffset(years=17)  # Rolling window
            new_start_date = new_start_date.strftime('%Y-01-01')
            
            cv_train_data = df[(df['date'] >= new_start_date) & (df['date'] < cv_year)]
            cv_test_data = df[df['year'] == cv_year[:4]]
            
            if not cv_test_data.empty and not cv_train_data.empty:
                # Train model with current parameters
                model = KNeighborsRegressor(**params)
                model.fit(cv_train_data[selected_columns], cv_train_data['Risk_Premium'])
                
                # Make predictions and calculate RMSE
                yhat_cv = model.predict(cv_test_data[selected_columns])
                rmse = np.sqrt(mean_squared_error(cv_test_data['Risk_Premium'], yhat_cv))
                cv_scores.append(rmse)
        
        if cv_scores:
            avg_score = np.mean(cv_scores)
            
            if avg_score < best_score:
                best_score = avg_score
                best_params = params.copy()
    
    print(f"Best parameters for {year}: {best_params}, CV Score: {best_score:.4f}")
    
    # Train and predict with best parameters for the current year
    current_year_data = test_data[test_data['year'] == year[:4]]
    
    if not current_year_data.empty:
        year_dt = pd.to_datetime(year)
        new_start_date = year_dt - pd.DateOffset(years=22)  # Rolling window
        new_start_date = new_start_date.strftime('%Y-01-01')
        
        train_data = df[(df['date'] >= new_start_date) & (df['date'] < year)]
        
        # Train model with best parameters
        model = KNeighborsRegressor(**best_params)
        model.fit(train_data[selected_columns], train_data['Risk_Premium'])
        
        # Make predictions
        yhat_oos = model.predict(current_year_data[selected_columns])
        
        # Store predictions and actual values
        for i in range(len(yhat_oos)):
            results.append({
                'month': current_year_data['date'].iloc[i],
                'predicted_return': yhat_oos[i],
                'actual_return': current_year_data['Risk_Premium'].iloc[i]
            })

# Convert results to DataFrame for analysis
results_df = pd.DataFrame(results)
```

```python
# Define the fixed investment amount
fixed_investment = 1000  # Amount to invest each month

# Create a DataFrame for cumulative returns
investment_results = []  # To store cumulative returns
cumulative_return = 0.0  # Start with zero cumulative return

# Loop through each unique month in the results
for month in sorted(results_df['month'].unique()):
    # Select the stocks for the current month
    current_month_stocks = results_df[results_df['month'] == month]

    if not current_month_stocks.empty:
        # Select the top 10% stocks based on predicted returns
        top_10_percent_threshold = current_month_stocks['predicted_return'].quantile(0.90)
        current_top_stocks = current_month_stocks[current_month_stocks['predicted_return'] >= top_10_percent_threshold]

        # Use the risk premium as the monthly return
        monthly_returns = current_top_stocks['actual_return'].values
        
        # Calculate the average risk premium of the selected top stocks
        if len(monthly_returns) > 0:
            average_monthly_return = monthly_returns.mean()  # Average risk premium
            
            # Calculate total return from the fixed investment
            cumulative_return += average_monthly_return * fixed_investment

            # Store the result for this month
            investment_results.append({
                'month': month,
                'cumulative_return': cumulative_return
            })

# Convert results to DataFrame
investment_df = pd.DataFrame(investment_results)

# Display cumulative returns
print(investment_df)
```

```python
# Compute cumulative portfolio value over time
investment_df['portfolio_value'] = investment_df['cumulative_return'] + fixed_investment  # Start with initial investment

# Calculate rolling max (highest portfolio value seen so far)
investment_df['peak_value'] = investment_df['portfolio_value'].cummax()

# Compute drawdown
investment_df['drawdown'] = (investment_df['portfolio_value'] - investment_df['peak_value']) / investment_df['peak_value']

# Compute max drawdown (worst drop from peak)
max_drawdown = investment_df['drawdown'].min()

# Display results
print(f"Maximum Drawdown: {max_drawdown:.2%}")
print(investment_df[['month', 'portfolio_value', 'peak_value', 'drawdown']])
```
